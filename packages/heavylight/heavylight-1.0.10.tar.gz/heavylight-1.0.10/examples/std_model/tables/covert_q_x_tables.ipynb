```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"qx_generic_raw.csv")
df
```

```python
df2 = df.melt(id_vars="Age exact x", value_vars=df.columns[1:], var_name="duration", value_name="q_x")
```

```python
df2["duration"] = df2["duration"].replace("5+", 5).astype(np.int64)
df2
```

```python
df2.dtypes
```

```python
df2.columns = ["age|int", "duration|int", "q_x|float"]   # add column type suffixes
df2
```

```python
df2_s = df2.copy() # smoker version
```

```python
df2_s["smoker_status|str"] = "S"
```

```python
df2["smoker_status|str"] = "N"
df2
```

```python
df2_s
```

```python
df2_s["q_x|float"] *= 2 # smoker rate = double normal rate
```

```python
df2_s
```

```python
df2_s["q_x|float"].max() # check <= 1
```

```python
df3 = pd.concat([df2, df2_s])
df3
```

```python
# sort the columns
df3 = df3[["age|int", "duration|int", "smoker_status|str", "q_x|float"]]
df3
```

```python
df3.to_csv("q_x_generic.csv", index=False)
```
