```python
import pandas as pd
import numpy as np

# Define the number of data points
n = 10000

# Generate random data for each distribution
data_uniform = np.random.uniform(low=0, high=100, size=n)
data_gaussian = np.random.normal(loc=50, scale=10, size=n)

data_poisson = np.random.poisson(lam=50, size=n)

# Create DataFrames
df_uniform = pd.DataFrame(data_uniform, columns=['x'])
df_gaussian = pd.DataFrame(data_gaussian, columns=['x'])
df_poisson = pd.DataFrame(data_poisson, columns=['x'])

# Create a dictionary of the DataFrames
dfs = {
    'uniform': df_uniform,
    'gaussian': df_gaussian,
    'poisson': df_poisson
}
```

```python
import matplotlib.pyplot as plt
from quickstats.plots import VariableDistributionPlot
plot_options = {
    'uniform': {
        # hide with custom function
        'hide': lambda x: (x < 10) | ((x > 40) & ( x < 50)) | (x > 80),
        'plot_format': 'hist',
        'error_format': 'fill'
    },
    'gaussian': {
        'error_format': 'errorbar'
    },
    'poisson': {
        # hide range
        'hide': (50, 60),
        'error_format': 'shade'
    }
}
plotter = VariableDistributionPlot(dfs, plot_options=plot_options)
```

```python
plotter.draw("x", show_error=True, stacked=False)
```

```python
plotter.draw("x", show_error=True, stacked=True)
```
