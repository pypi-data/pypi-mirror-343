```python
import numpy as np
import pandas as pd

from quickstats.plots import MultiPanel1DPlot

size = 10
dfs = {
    'A': pd.DataFrame({
        'x': np.random.rand(size),
        'y': np.random.rand(size)
    }),
    'B': pd.DataFrame({
        'x': np.random.rand(size),
        'y': np.random.rand(size)
    }),
    'C': pd.DataFrame({
        'x': np.random.rand(size),
        'y': np.random.rand(size)
    })
}
```

```python
plotter = MultiPanel1DPlot(dfs)
```

```python
plotter.draw(
    'x', 'y',
    xlabel='X',
    ylabels=['Y(A)', 'Y(B)', 'Y(C)'],
    panel_targets=['A', 'B', 'C'],
    ypads=[(0.2, 0.3), (0.2, 0.3), (0.2, 0.3)],
)
```
