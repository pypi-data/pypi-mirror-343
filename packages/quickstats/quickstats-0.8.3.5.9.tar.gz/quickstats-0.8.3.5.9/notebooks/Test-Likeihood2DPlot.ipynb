```python
from quickstats.plots import Likelihood2DPlot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create some data
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = - (Z1 - Z2) * 2
Z = Z - np.min(Z)

X, Y, Z = X.flatten(), Y.flatten(), Z.flatten()
df = pd.DataFrame({'x': X, 'y': Y, 'z': Z})
```

```python
plotter = Likelihood2DPlot(df, styles={
   # 'legend': {'handleheight': 1},
    'bestfit': {'marker': '*'},
    'contour': {'linestyles': '-'},
    'contourf': {'hatches': ['//', '-', '']} },
    color_cycle='default')
```

```python
plotter.add_point(0, 0, label='Test', name='Test', marker= 'o', color='r', markerfacecolor='none')
```

```python
plotter.draw('x', 'y', 'z',
             draw_contour=True,
             draw_contourf=True,
             confidence_levels=(0.2, 0.5, 0.77), sigma_levels=None)
```
