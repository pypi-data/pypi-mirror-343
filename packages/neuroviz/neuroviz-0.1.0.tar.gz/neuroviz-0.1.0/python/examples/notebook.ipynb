```python
from neuroviz import NeuroViz, ParameterDict

neuro = NeuroViz(port=9001, use_secret=False)
```

```python
neuro.set_live_parameters({
  'outline': 0.1,
  'see_through': 0.5,
  'smoothness': 0.1,
  'transparency': 0.2
})
```

```python
a: ParameterDict = {
  'outline': 0.1,
  'see_through': 0.5,
  'smoothness': 0.1,
  'transparency': 0.2
}

b = {
  'outline': 0.5,
  'see_through': 0.2,
  'smoothness': 0.5,  
  'transparency': 0.1
}

chosen = neuro.prompt_choice(a, b)
print(chosen)
```

```python
a: ParameterDict = {
  'outline': 0.1,
  'see_through': 0.5,
  'smoothness': 0.1,
  'transparency': 0.2
}

rating = chosen = neuro.prompt_rating(a)
print(rating)
```
