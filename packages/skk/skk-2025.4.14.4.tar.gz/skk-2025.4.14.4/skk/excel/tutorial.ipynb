## 第 2 步

编写计算方式

示例:

```python
c = 表格.列

c.已获利润 = (c.售价 - c.进价) * c.实际销量

c.剩余库存 = c.进货数量 - c.实际销量

c.是否完成销售目标 = 判断(c.实际销量 >= c.计划销量, '是', '否')

c.进货建议 = 判断(c.剩余库存 < 100, '须进货', '无须进货')  # 假设库存小于100时, 须进货
```

在下面编写实际的计算方式：

````python
c = 表格.列

c.已获利润 = (c.售价 - c.进价) * c.实际销量

c.剩余库存 = c.进货数量 - c.实际销量

c.是否完成销售目标 = 判断(c.实际销量 >= c.计划销量, '是', '否')

c.进货建议 = 判断(c.剩余库存 < 100, '须进货', '无须进货')
````
